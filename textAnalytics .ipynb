```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
```

```python
text= "Machine learning is very interesting and useful for data analysis."
```

```python
tokens=word_tokenize(text)
print(tokens)
```

```python
pos= pos_tag(tokens)
print(pos)
```

```python
stop_words= set( stopwords.words('english'))
filtered= [ w for w in tokens if w.lower() not in stop_words]
print("After stop word removal:", filtered)
```

```python
ps= PorterStemmer()
stemmed= [ps.stem(w) for w in filtered]
print("Stemmed words:", stemmed)
```

```python
lm= WordNetLemmatizer()
lemmatized= [lm.lemmatize(w) for w in filtered]
print(lemmatized)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
doc = [
  "machine learning is fun",
  "learning is powerful"
]
cv = CountVectorizer()
tf= cv.fit_transform(doc)
print(cv.get_feature_names_out())
```

```python
print(tf.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf= TfidfVectorizer()
result= tfidf.fit_transform(doc)
print("words:",tfidf.get_feature_names_out())
print("TF-IDF:",result.toarray())
```
